## Problem 1

```python
from scipy import stats
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))

for i in range(9):
    ax = fig.add_subplot(3,3,i+1)
    n = 10*2**i
    ax.hist(stats.norm.rvs(size=n))
    ax.set_title('n={}'.format(n))
    
plt.tight_layout()
fig.savefig('q1_p1')
plt.show()
```

## Problem 2

```python
fig = plt.figure(figsize=(6,12))

n = 1500
distributions = [stats.norm.rvs(loc=0, scale=0.5, size=n),
                 stats.norm.rvs(loc=0, scale=2, size=n),
                stats.t.rvs(df=2, size=n),
                stats.t.rvs(df=20, size=n),
                stats.expon.rvs(scale=1/0.25, size=n),
                stats.expon.rvs(scale=1/4, size=n),
                stats.beta.rvs(a=10, b=90, size=n),
                stats.beta.rvs(a=20, b=80, size=n)]
labels = ['Normal(0,0.5)',
         'Normal(0,2)',
         'Student(2)',
         'Student(20)',
         'Exponential(0.25)',
         'Exponential(4)',
         'Beta(10,90)',
         'Beta(20,80)']

for i in range(8):
    ax = fig.add_subplot(4,2,i+1)
    ax.hist(distributions[i])
    ax.set_title(labels[i])
    
plt.tight_layout()
fig.savefig('q1_p2_hist')
plt.show()
```

```python
fig = plt.figure(figsize=(6,12))

for i in range(8):
    ax = fig.add_subplot(4,2,i+1)
    stats.probplot(x=distributions[i],
                   sparams=(0,1),
                   dist=stats.norm,
                   plot=ax)
    ax.set_title(labels[i])
    
plt.tight_layout()
fig.savefig('q1_p2_qqplot')
plt.show()
```
