```python
import json
import requests

from bs4 import BeautifulSoup


url = 'http://tcpip.epfl.ch/output.php'


stats_recieved = []
for i in range(1,1000) :
    c = str(i)
    payload = {"sciper": '217575', "clients": c, "apoints": '1', "servers": '1'}
    r = requests.post(url, data=payload)

    soup = BeautifulSoup(r.content, 'html.parser')
    tr = soup.find_all('tr')[6:]

    results  = []
    for values in tr :
        results.append(values.find_all('td')[1])

    final_results = []
    for r in results :
        final_results.append(float(r.text))

    stats_recieved.append(final_results)

print(stats_recieved)
```

the choice of plot we choose is a scatter plot

```python
C = np.arange(1, 1000, 1)
```

```python
theta = []
packets_per_second = []
collision_probablity = []
delay = []

for values in stats_recieved :
    theta.append(values[0])
    packets_per_second.append(values[1])
    collision_probablity.append(values[2])
    delay.append(values[3])
```

```python
def plot(C,y,y_label):
    plt.scatter(C,y)
    plt.xlabel("C")
    plt.ylabel(y_label)
    plt.show()
```

```python
plot(C, theta, 'theta')
```

```python
plot(C, packets_per_second, 'packets per second')
```

```python
plot(C, collision_probablity, 'collistion probablity')
```

```python
plot(C, delay, 'delay')
```
